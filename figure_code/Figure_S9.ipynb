```python
fig = plt.figure(figsize=(8,10))

ax1 = fig.add_subplot(3,1,1)
ax1.set_title('LM', fontsize=14)
index = ['Fast', 'Slow', 'Passive']
Y_f = [average(LM_1_f), average(LM_2_f), average(LM_3_f)]
Y_l = [average(LM_1_l), average(LM_2_l), average(LM_3_l)]
error_f = [LM_1_f.std(), LM_2_f.std(), LM_3_f.std()]
error_l = [LM_1_l.std(), LM_2_l.std(), LM_3_l.std()]
bar_width = 0.3
ax1.set_ylabel('Carbon mean age (yr)', fontsize=14)
ax1.bar(arange(len(index)) - bar_width/2, Y_f, yerr=error_f, width=0.3, color = material['blue'][7], 
           error_kw={'ecolor': '0.2', 'capsize': 2}, label='steady state')
ax1.bar(arange(len(index)) + bar_width/2, Y_l, yerr=error_l, width=0.3, color = material['blue'][2], hatch='//',
           error_kw={'ecolor': '0.2', 'capsize': 2}, label='last year')
ax1.text(0.02,0.9,'(a)', fontsize=15, transform=ax1.transAxes)
ax1.set_xticks(arange(len(index)))
ax1.set_xticklabels(index)
ax1.tick_params('both', labelsize=14)
ax1.legend(loc=(0.1,0.76))
ax1.set_yticks([0,20,40,60])

ax2 = fig.add_subplot(3,1,2)
ax2.set_title('rMM', fontsize=14)
index = ['Litter', 'Soil', 'Microbe']
Y_f = [average(rMM_l_f), average(rMM_s_f), average(rMM_b_f)]
Y_l = [average(rMM_l_l), average(rMM_s_l), average(rMM_b_l)]
error_f = [rMM_l_f.std(), rMM_s_f.std(), rMM_b_f.std()]
error_l = [rMM_l_l.std(), rMM_s_l.std(), rMM_b_l.std()]
bar_width = 0.3
ax2.set_ylabel('Carbon mean age (yr)', fontsize=14)
ax2.bar(arange(len(index)) - bar_width/2, Y_f, yerr=error_f, width=0.3, color = material['blue'][7], 
           error_kw={'ecolor': '0.2', 'capsize': 2}, label='steady state')
ax2.bar(arange(len(index)) + bar_width/2, Y_l, yerr=error_l, width=0.3, color = material['blue'][2], hatch='//',
           error_kw={'ecolor': '0.2', 'capsize': 2}, label='last year')
ax2.text(0.02,0.9,'(b)', fontsize=15, transform=ax2.transAxes)
ax2.set_xticks(arange(len(index)))
ax2.set_xticklabels(index)
ax2.tick_params('both', labelsize=14)
ax2.set_yticks([0,50,100,150])

ax3 = fig.add_subplot(3,1,3)
ax3.set_title('MM', fontsize=14)
index = ['Litter', 'Soil', 'Microbe']
Y_f = [average(MM_l_f), average(MM_s_f), average(MM_b_f)]
Y_l = [average(MM_l_l), average(MM_s_l), average(MM_b_l)]
error_f = [MM_l_f.std(), MM_s_f.std(), MM_b_f.std()]
error_l = [MM_l_l.std(), MM_s_l.std(), MM_b_l.std()]
bar_width = 0.3
ax3.set_ylabel('Carbon mean age (yr)', fontsize=14)
ax3.bar(arange(len(index)) - bar_width/2, Y_f, yerr=error_f, width=0.3, color = material['blue'][7], 
           error_kw={'ecolor': '0.2', 'capsize': 2}, label='steady state')
ax3.bar(arange(len(index)) + bar_width/2, Y_l, yerr=error_l, width=0.3, color = material['blue'][2], hatch='//',
           error_kw={'ecolor': '0.2', 'capsize': 2}, label='last year')
ax3.text(0.02,0.9,'(c)', fontsize=15, transform=ax3.transAxes)
ax3.set_xticks(arange(len(index)))
ax3.set_xticklabels(index)
ax3.tick_params('both', labelsize=14)
ax3.set_yticks([0,50,100,150])

plt.subplots_adjust(hspace=0.4, wspace=0.25)
plt.savefig('Figure. S10.png', dpi = 600, bbox_inches='tight')
```
